Dificuldade na Previsão: Analisar uma doença nova também significa que é mais difícil fazer previsões precisas. A falta de dados históricos sólidos pode limitar a capacidade de entender completamente o comportamento da doença.

Para superar essas dificuldades, é essencial ser transparente sobre a origem dos dados, considerar as limitações existentes e estar ciente de que as informações podem mudar à medida que mais é aprendido sobre a doença. ## **Análise Exploratória para o Brasil**

```python
# data frame apenas com dados do Brasil
data_brasil = data.loc[data.location == "Brazil"].copy()
```

```python
data_brasil
```

```python

# identificaçãso do primeiro caso
data_brasil.loc[data_brasil['new_cases'] != 0.0].head(1)
print('O primeiro caso foi registrado em {}'.format(data_brasil.date.loc[3154]))
```

```python
# Primeira morte no brasil por covid

# Filtrar os dados para o Brasil e mortes maiores que zero
primeira_morte_brasil = data[(data['location'] == 'Brazil') & (data['total_deaths'] > 0)]

# Encontrar a data da primeira morte
data_primeira_morte = primeira_morte_brasil['date'].min()

print(f'A primeira morte por COVID-19 no Brasil foi registrada em {data_primeira_morte}.')
```

#Uma informação interessante para investigarmos é quanto tempo levou desde o primeiro caso até a primeira morte.


```python

# Filtrar os dados para o Brasil e casos maiores que zero
casos_brasil = data[(data['location'] == 'Brazil') & (data['total_cases'] > 0)]

# Encontrar a data do primeiro caso
data_primeiro_caso = casos_brasil['date'].min()

# Filtrar os dados para o Brasil e mortes maiores que zero a partir da data do primeiro caso
mortes_apos_primeiro_caso = data[(data['location'] == 'Brazil') & (data['total_deaths'] > 0) & (data['date'] >= data_primeiro_caso)]

# Encontrar a data da primeira morte após o primeiro caso
data_primeira_morte_apos_primeiro_caso = mortes_apos_primeiro_caso['date'].min()

# Calcular a diferença em dias
diferenca_dias = (pd.to_datetime(data_primeira_morte_apos_primeiro_caso) - pd.to_datetime(data_primeiro_caso)).days

print(f'Demorou {diferenca_dias} dias para o Brasil registrar a primeira morte após o primeiro caso de COVID-19.')
```

# Demorou 21 dias para o Brasil registrar a primeira morte após o primeiro caso de COVID-19.

Por fim, vamos plotar gráficos de linha para compararmos a evolução dos casos e mortes no Brasil.


```python
# Plotar gráficos de linha em escalas linear e logarítmica para os casos totais e mortes totais.
x = list(range(168))
z = data_brasil['total_cases']
z2 = data_brasil['total_deaths']

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(9, 7))
ax1.set_title('Casos no Brasil')
ax1.set_xlabel('Dias')
ax1.set_ylabel('Casos')
ax1.plot(x, z)

ax2.set_title('Casos no Brasil (Escala logarítmica)')
ax2.set_xlabel('Dias')
ax2.set_ylabel('Casos')
ax2.set_yscale('log')
ax2.plot(x, z)

ax3.set_title('Mortes no Brasil')
ax3.set_xlabel('Dias')
ax3.set_ylabel('Mortes')
ax3.plot(x, z2)

ax4.set_title('Mortes no Brasil (Escala logarítmica')
ax4.set_xlabel('Dias')
ax4.set_ylabel('Mortes')
ax4.set_yscale('log')
ax4.plot(x, z2)
plt.tight_layout()
```
